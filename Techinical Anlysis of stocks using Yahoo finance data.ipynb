## Step 2: 
Analyze the dataset to gather the following insights:

1. Identify the data type of each column.
2. Determine the total count of elements in each column.
3. Calculate the overall count of missing values in each column.
4. Compute the percentage of missing values in the dataset.


```python
def get_stock_insights(dataframe):
    stock_insights = pd.DataFrame({
        'Datatype': dataframe.dtypes, # Data types of the columns
        'Total_Elements': dataframe.count(), # Total elements in the columns
        'Null_Count': dataframe.isnull().sum(), # Total null values inthe  columns
        'Null_Percentage': dataframe.isnull().sum() / len(dataframe) * 100 # Percentage of the null values
    })
    return stock_insights
```

## Step 4: Market Capitalization


This provides the valuation of a company that is actively traded in the stock market. This enables investors to gauge the perceived worth of the company in the eyes of the investment community.

The magnitude and valuation of a company indicate the potential level of risk associated with investing in its stock, as well as the possible returns over time.

This valuation is derived by multiplying the total count of outstanding shares with the current price per share.

Market Capitalization (M_Cap) = Total Number of Shares * Current Share Price

### Identify Stocks Offering Enhanced Returns with Reduced Risk Across Time
Ascertain the market capitalization figures for each of the three stocks and visualize the outcomes to deduce which stocks offer a combination of decreased risk and amplified returns over time.

```python
apple['M_Cap'] = apple['Open'] * apple['Volume']
tesla['M_Cap'] = tesla['Open'] * tesla['Volume']
amazon['M_Cap'] = amazon['Open'] * amazon['Volume']
```

```python
apple['M_Cap'].plot(label = 'Apple', figsize = (15,7))
tesla['M_Cap'].plot(label = 'Tesla')
amazon['M_Cap'].plot(label = 'Amazon')

plt.title('Market Cap')
plt.legend()
```

####  Conclusion: Tesla is traded more than apple and amazon, so is we only look at the numbers than buying tesla would be safest. Also, apple is a bit safer than amazon.

## Step 5: Moving Averages:

Utilizing Moving Averages to Address Price Fluctuations, as Stock Prices Exhibit High Volatility and Rapid Changes Over Time. Typically, an upward trajectory in a moving average (indicating a bull market) could signify an uptick in a security's price or momentum, whereas a downward trend (indicative of a bear market) might reflect a decline.

We can leverage the utility of the 50-day and 200-day averages to detect trends. The 200-day moving average holds particular significance in stock trading. As long as a stock price's 50-day moving average remains higher than its 200-day moving average, the stock is generally perceived to be experiencing a bullish trend.

#### Determine the Stocks that Have Significant Upward Trends

Find the Moving Averages for each of the three stocks and compare them to find which stocks have significant bullish trends. 


```python
# Moving avg of just one stock: apple

apple['MA50'] = apple['Open'].rolling(50).mean()
apple['MA200'] = apple['Open'].rolling(200).mean()

# Plot

apple['Open'].plot(figsize = (15,7))
apple['MA50'].plot()
apple['MA200'].plot()
plt.legend()
```

```python
apple['MA50'] = apple['Open'].rolling(50).mean()
apple['MA200'] = apple['Open'].rolling(200).mean()

amazon['MA50'] = amazon['Open'].rolling(50).mean()
amazon['MA200'] = amazon['Open'].rolling(200).mean()

tesla['MA50'] = tesla['Open'].rolling(50).mean()
tesla['MA200'] = tesla['Open'].rolling(200).mean()
```

```python
# plot

figure, axes = plt.subplots(2, 2, figsize = (15, 10))

figure.suptitle('Moving Averages for Apple, Tesla and Amazon')

axes[0,0].set_title('Apple')
axes[0,1].set_title('tesla')
axes[1,0].set_title('Amazon')

apple['MA50'].plot(ax=axes[0, 0])
apple['MA200'].plot(ax=axes[0, 0])
apple['Open'].plot(ax=axes[0, 0])

tesla['MA50'].plot(ax=axes[0, 1])
tesla['MA200'].plot(ax=axes[0, 1])
tesla['Open'].plot(ax=axes[0, 1])

amazon['MA50'].plot(ax=axes[1, 0])
amazon['MA200'].plot(ax=axes[1, 0])
amazon['Open'].plot(ax=axes[1, 0])
```

#### Conclusion: 
1. Apple has a relatively consistent upward trend, with a slight decrease towards the end before increasing again.
2. Tesla has a relatively consistent upward trend, with consistent decrease in 2022.
3. Amazon has a sharp increase, followed by a period of an almost consistent trend. There is a sharp drop towards the end before it rises again. 

## Step 6: Volatility and Stability

A percentage rise in stock value indicates the variation in stock price compared to the previous day. The magnitude of this change, whether positive or negative, directly correlates with the stock's level of volatility.

The formula used is:

rt = (Pt / Pt-1) - 1

Where:
- rt is the Return
- Pt is the Price at time T
- Pt-1 is the Price at time T-1

Investors often seek stability and tend to be apprehensive about volatility.

By employing histograms, we can assess whether a stock exhibits stability or volatility. A denser histogram indicates a more volatile nature of the stock.

#### Compare Volatility and Stability

Find and compare the volatility and stability for each of the 3 stocks and determine which stock appears to be the most volatile.

```python
apple['returns'] = (apple['Close']/apple['Close'].shift(1)) -1
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) -1
amazon['returns'] = (amazon['Close']/amazon['Close'].shift(1)) -1

#Draw 
figure, axes = plt.subplots(2, 2, figsize = (15, 10))

figure.suptitle('Stability and Volatility for Apple, Amazon, Ford, and Cisco')

axes[0,0].set_title('Apple')
axes[0,1].set_title('Tesla')
axes[1,0].set_title('Amazon')

apple['returns'].hist(bins = 100, label = 'Apple', alpha = 0.5, ax=axes[0, 0])

tesla['returns'].hist(bins = 100, label = 'Tesla', alpha = 0.5, ax=axes[0, 1])

amazon['returns'].hist(bins = 100, label = 'Amazon', alpha = 0.5, ax=axes[1, 0])
```

```python
# Comparing the volatility / stability for all the stocks.

apple['returns'] = (apple['Close']/apple['Close'].shift(1)) -1
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) -1
amazon['returns'] = (amazon['Close']/amazon['Close'].shift(1))-1

apple['returns'].hist(bins = 100, label = 'Apple', alpha = 0.5, figsize = (15,7))
tesla['returns'].hist(bins = 100, label = 'Tesla', alpha = 0.5)
amazon['returns'].hist(bins = 100, label = 'Amazon', alpha = 0.5)

plt.legend()
```

#### Conclusion:  Tesla seems to be most volatile as its histogram is thickest.
